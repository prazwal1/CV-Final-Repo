```python
import torchvision.datasets as dset
import torchvision.transforms as T

# Define the root directory where the data will be stored
root_dir = 'VOC' 

# Download VOC 2012 Training and Validation sets
voc_trainval = dset.VOCDetection(
    root=root_dir,
    year='2012', 
    image_set='trainval', # Downloads both train and val splits
    transform=T.ToTensor()
)

# Download VOC 2007 Test set
voc_test = dset.VOCDetection(
    root=root_dir,
    year='2007',
    image_set='test',
    transform=T.ToTensor()
)

print(f"VOC 2012 trainval samples: {len(voc_trainval)}")
print(f"VOC 2007 test samples: {len(voc_test)}")
```
